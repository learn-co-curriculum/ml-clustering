# K means clustering

The k means clustering algorithm is an iterative algorithm that rearches for a pre-determined number of clusters within an unlabeled dataset, and basically works as follows:

- select k initial seeds
- assign each observation to the clusted to which it is "closest" 
- recompute the cluster centroids
- reassign the observations to one of the clusters according to some rule
- stop if there is no reallocation

### 1.1 Transform strings to numeric values

LORE DO only create this function when using an _actual_ data set.

This function transforms strings to numeric values, which is necessary as performing a clustering algorithm all about computing distances.

```python
def from_str(dataset, attributes):
    """
    input: The dataset name and the attribute name (or attribute names as a lists) for attributes that are strings.
    output: Transformed data    
    """
    # convert when it's a list
    if type(attributes) == list:
        for attr_name in attributes:
            dataset[attr_name] = dataset[attr_name].astype(float)

    # convert when there's just one
    else:
        data[attributes] = data[attributes].astype(float)
    
    # Return dataset after conversion
    return dataset
```

### 1.2 initialize cluster centroids

The function `initialization` generates random cluster centroids.

- You'll create `n_attr` first. Use the shape of `data` to store the number of attributes here.
- create a matrix with the dimensions (number of clusters, number of attributes) and store it in `centroids`. Use `np.mat` along with np.zeros`
- Next, you'll make sure random centroids are created within the bounds of the data.

Now, use the function on `X` make sure to create 7 cluster centroids. store the results in `init`

**ORIGINAL**

```python
# from math import sqrt, floor
def initialize(data, k):

    """    
    input:  data - ndarray, dataset to cluster instances from
            k - number of clusters
    output: ndarray - k centroids as numpy ndarray
    """
    # store the number of attributes, and create an empty centroids vector
    n_attr = data.shape[1]
    centroids = np.mat(np.zeros((k, n_attr))) 
    
    # Create random centroids (get min, max attribute values, randomize in that range)
    for i in range(n_attr):
        min_i = min(data[:,i])
        max_i = max(data[:,i])
        centroids[:,i] = np.random.rand(k, 1) * (max_i - min_i) + min_i
    return centroids
```

```python
# from math import sqrt, floor
def initialize(data, k):

    """    
    input:  data - ndarray, dataset to cluster instances from
            k - number of clusters
    output: ndarray - k centroids as numpy ndarray
    """
    # store the number of attributes, and create an empty centroids vector
    n_attr = data.shape[1] 
    centroids = np.zeros((n_attr, k))
    
    # Create random centroids (get min, max attribute values, randomize in that range)
    for i in range(n_attr):
        min_i = min(data[:,i])
        max_i = max(data[:,i])
        centroids[i,:] = np.squeeze(np.random.rand(k, 1))* (max_i - min_i) + min_i
    return centroids.T
```

### 1.4 Perform cluster assignments

```python
import numpy as np

def cluster(ds, k):

    '''
    The k-means clustering algorithm
    
    Parameters:
    -----------
    ds: ndarray
        Dataset to cluster instances from
    k: int
        The number of clusters
    Returns:
    --------
    ndarray
        Resulting centroids after clustering
    ndarray
        Cluster assignments after clustering
    int
        Number of iterations required by clustering algorithm
    ndarray
        Original centroids
    '''

    # Number of rows in dataset
    m = np.shape(ds)[0]

    # Hold the instance cluster assignments
    cluster_assignments = np.mat(np.zeros((m, 2)))

    # Initialize centroids
    cents = initialize(ds, k)
    
    # Preserve original centroids
    cents_orig = cents.copy()
    
    changed = True
    num_iter = 0

    # Loop until no changes to cluster assignments
    while changed:

        changed = False

        # For every instance (row in dataset)
        for i in range(m):

            # Track minimum distance, and vector index of associated cluster
            min_dist = np.inf
            min_index = -1

            # Calculate distances
            for j in range(k):

                dist_ji = dist.euclidian(cents[j,:], ds[i,:])
                if dist_ji < min_dist:
                    min_dist = dist_ji
                    min_index = j

            # Check if cluster assignment of instance has changed
            if cluster_assignments[i, 0] != min_index: 
                changed = True

            # Assign instance to appropriate cluster
            cluster_assignments[i, :] = min_index, min_dist**2

        # Update centroid location
        for cent in range(k):
            points = ds[np.nonzero(cluster_assignments[:,0].A==cent)[0]]
            cents[cent,:] = np.mean(points, axis=0)

        # Count iterations
        num_iter += 1

    # Return important stuff when done
    return cents, cluster_assignments, num_iter, cents_orig
```
